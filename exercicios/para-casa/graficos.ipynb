# Tratando a base de dados

```python
df_amazon_reduced = df_amazon[['discounted_price','actual_price','discount_percentage','rating','rating_count']]
df_amazon_reduced
```

```python
df_amazon_reduced['discounted_price'] = df_amazon['discounted_price'].str.replace('₹', '', regex=False)
df_amazon_reduced
```

```python
df_amazon_reduced['actual_price'] = df_amazon['actual_price'].str.replace('₹', '', regex=False)
df_amazon_reduced	
```

```python
df_amazon_reduced['discount_percentage'] = df_amazon['discount_percentage'].str.replace('%', '', regex=False)
df_amazon_reduced	
```

```python
df_amazon_reduced.describe()
```

```python
df_amazon_reduced.info()
```

```python
object_columns = ['discounted_price','actual_price','discount_percentage','rating','rating_count']
object_columns
```

```python
for col in object_columns:
    df_amazon_reduced[col] = pd.to_numeric(df_amazon_reduced[col], errors='coerce')
    if df_amazon_reduced[col].isnull().any():
        df_amazon_reduced[col].fillna(0, inplace=True)  
    
    if col == 'rating':  
        df_amazon_reduced[col] = df_amazon_reduced[col].astype(float)
    else:  
        df_amazon_reduced[col] = df_amazon_reduced[col].astype(int)
```

```python
df_amazon_reduced.info()
```

# Gráfico 1

```python
plt.figure(figsize=(10, 5))
plt.hist(df_amazon_reduced['discounted_price'], bins=10, color='blue', edgecolor='black')
plt.title('Histograma de Discounted Price')
plt.xlabel('Discounted Price')
plt.ylabel('Frequência')
plt.grid(True)
plt.show()
```

# Gráfico 2

```python
plt.figure(figsize=(10, 5))
plt.hist(df_amazon_reduced['actual_price'], bins=10, color='green', edgecolor='black')
plt.title('Histograma de Actual Price')
plt.xlabel('Actual Price')
plt.ylabel('Frequência')
plt.grid(True)
plt.show()
```

# Gráfico 3

```python
plt.figure(figsize=(10, 5))
plt.scatter(df_amazon_reduced['discount_percentage'], df_amazon_reduced['rating'], color='red')
plt.title('Discount Percentage vs Rating')
plt.xlabel('Discount Percentage')
plt.ylabel('Rating')
plt.grid(True)
plt.show()
```

# Gráfico 4

```python
plt.figure(figsize=(10, 5))
plt.scatter(df_amazon_reduced['rating_count'], df_amazon_reduced['rating'], color='purple')
plt.title('Rating Count vs Rating')
plt.xlabel('Rating Count')
plt.ylabel('Rating')
plt.grid(True)
plt.show()
```
